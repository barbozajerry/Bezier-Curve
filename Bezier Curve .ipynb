```python
def Bezier(t,A): 
    a0, a1, a2, a3 = A[0], 3*(A[2]-A[0]), 3*(A[0]+A[4]-2*A[2]), A[6] -A[0] + 3*A[2] - 3*A[4]
    b0, b1, b2, b3 = A[1], 3*(A[3]-A[1]), 3*(A[1]+A[5]-2*A[3]), A[7] -A[1] + 3*A[3] - 3*A[5]
    return [a0+a1*t+a2*t**2+a3*t**3, b0+b1*t+b2*t**2+b3*t**3]
```

```python
for i in range(M.shape[0]): 
    # plot Bezier curve i
    t = np.linspace(0,1,10)
    B = Bezier(t,M[i,:])
    plt.plot(B[0],B[1])
    # plot the left end point i 
    plt.plot(M[i,0],M[i,1], 'r*')
    # plot the right end point i+1
    plt.plot(M[i,6],M[i,7], 'b*')
    # plot the left controlpoint i+
    plt.plot(M[i,2],M[i,3], 'c*')
    # plot the left controlpoint i+
    plt.plot(M[i,4],M[i,5], 'g*')
plt.show()
    
    
```
